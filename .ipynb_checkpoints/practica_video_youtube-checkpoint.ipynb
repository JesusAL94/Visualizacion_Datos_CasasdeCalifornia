### Precio de venta vs calidad
Relaciones categóricas --gráficas de caja y bigote o boxplot

```python
var = 'OverallQual'
```

```python
data = pd.concat([df_train['SalePrice'], df_train[var]], axis = 1)
```

```python
f, ax = plt.subplots(figsize = (8,6)) #f y ax son las variables que siempre se utilizar para definir una figura
# esto esta preparado para introducir una gráfica.
```

```python
f, ax = plt.subplots(figsize = (8,6))
fig = sns.boxplot(x=var, y='SalePrice', data = data)
fig.axis(ymin=0, ymax=800000)
```

Fig Gráfico de cajas: claramente un crecimiento exponencial, donde a mayor calidad, mayor disparidad de precios(la amplitud de las cajas)

### Precio de venta vs construcción

```python
var = 'YearBuilt'
data = pd.concat([df_train['SalePrice'], df_train[var]], axis=1)
f, ax = plt.subplots(figsize = (20,8))
fig = sns.boxplot(x=var, y='SalePrice', data = data)
fig.axis(ymin = 0, ymax = 800000)
plt.xticks(rotation=90)
plt.show()
```

### Heatmap de correccion
- Matriz de correlación

```python
print(df_train.dtypes) #para ver el tipo que son cada columna
```

```python
corrmat = df_train.select_dtypes(include=['number']).corr() #corrmat --> correlation matrix - matriz de correlación,
# esto es para hacer la correlacion solo con números
```

```python
f, ax = plt.subplots(figsize=(12,9))
sns.heatmap(corrmat, vmax=0.8, square = True)
plt.show()
```

### Heatmap pero con datos numéricos
- Correlacion con datos numéricos (igual pero mostrando los datos encima de los cuadrados)

```python
k = 10
cols = corrmat.nlargest(k, 'SalePrice')['SalePrice'].index 
```

```python
cols
```

```python
cm = np.corrcoef(df_train[cols].values.T) # De la biblioteca NumPy, utilizar el coeficiente de correlación para arreglos
```

```python
cm
```

```python
hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='0.2f', annot_kws={'size':10,}, yticklabels=cols.values, xticklabels=cols.values)
plt.show()
```
